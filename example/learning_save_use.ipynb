```python
import numpy as np
import pandas as pd

# Преобразуем data_y в одномерный массив
if isinstance(data_y, pd.DataFrame):  # Если data_y — DataFrame
    data_y = data_y.values.ravel()  # Преобразуем в массив NumPy и делаем его одномерным
elif isinstance(data_y, np.ndarray):  # Если data_y — массив NumPy
    if len(data_y.shape) > 1 and data_y.shape[1] == 1:  # Если форма (n_samples, 1)
        data_y = data_y.ravel()  # Преобразуем в одномерный массив
elif isinstance(data_y, list):  # Если data_y — список
    data_y = np.array(data_y).ravel()  # Преобразуем в массив NumPy и делаем его одномерным
else:
    raise ValueError("Unsupported data type for data_y. Please provide a 1D array-like object.")

# Теперь создаем Series
data_y = pd.Series(data_y)

# Разделим данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(data_X, data_y, test_size=0.2, random_state=42)
```

## 3. Испытанные алгоритмы сравнены по точности, а также по скорости работы

```python
# 1. Линейная регрессия
linear_model = LinearRegression()
start_train = time.time()
linear_model.fit(X_train, y_train)
end_train = time.time()
start_predict = time.time()
y_pred_linear = linear_model.predict(X_test)
end_predict = time.time()
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
print(f"Linear Regression - MSE: {mse_linear:.4f}, R2: {r2_linear:.4f}")
print(f"Training time: {end_train - start_train:.4f} seconds")
print(f"Prediction time: {end_predict - start_predict:.4f} seconds\n")
```

```python
# 2. Дерево решений для регрессии
tree_model = DecisionTreeRegressor(random_state=42)
start_train = time.time()
tree_model.fit(X_train, y_train)
end_train = time.time()
start_predict = time.time()
y_pred_tree = tree_model.predict(X_test)
end_predict = time.time()
mse_tree = mean_squared_error(y_test, y_pred_tree)
r2_tree = r2_score(y_test, y_pred_tree)
print(f"Decision Tree - MSE: {mse_tree:.4f}, R2: {r2_tree:.4f}")
print(f"Training time: {end_train - start_train:.4f} seconds")
print(f"Prediction time: {end_predict - start_predict:.4f} seconds\n")
```

```python
# 3. Градиентный бустинг (XGBoost)
xgb_model = XGBRegressor(random_state=42)
start_train = time.time()
xgb_model.fit(X_train, y_train)
end_train = time.time()
start_predict = time.time()
y_pred_xgb = xgb_model.predict(X_test)
end_predict = time.time()
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)
print(f"XGBoost - MSE: {mse_xgb:.4f}, R2: {r2_xgb:.4f}")
print(f"Training time: {end_train - start_train:.4f} seconds")
print(f"Prediction time: {end_predict - start_predict:.4f} seconds\n")
```

```python

# Функция для вычисления метрик
def evaluate_model(y_true, y_pred):
    """
    Вычисляет R2 и RMSE для предсказаний модели.
    :param y_true: истинные значения
    :param y_pred: предсказанные значения
    :return: R2 и RMSE
    """
    r2 = r2_score(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return r2, rmse

# Пример использования для трех моделей
metrics = []

# 1. Линейная регрессия
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)
r2_linear, rmse_linear = evaluate_model(y_test, y_pred_linear)
metrics.append(('Linear Regression', r2_linear, rmse_linear))

# 2. Дерево решений
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)
r2_tree, rmse_tree = evaluate_model(y_test, y_pred_tree)
metrics.append(('Decision Tree', r2_tree, rmse_tree))

# 3. Градиентный бустинг (XGBoost)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
r2_xgb, rmse_xgb = evaluate_model(y_test, y_pred_xgb)
metrics.append(('XGBoost', r2_xgb, rmse_xgb))

# Вывод результатов
print(f"{'Model':<20} {'R2':<10} {'RMSE':<10}")
print("-" * 40)
for model_name, r2, rmse in metrics:
    print(f"{model_name:<20} {r2:<10.4f} {rmse:<10.4f}")

# Вычисление средних значений
average_r2 = np.mean([metric[1] for metric in metrics])
average_rmse = np.mean([metric[2] for metric in metrics])

print("\nAverage Metrics:")
print(f"Average R2: {average_r2:.4f}")
print(f"Average RMSE: {average_rmse:.4f}")
```

```python
def regression_accuracy(y_true, y_pred, tolerance=0.1):
    """
    Вычисляет точность для регрессии как долю предсказаний,
    которые находятся в пределах допустимой погрешности.
    :param y_true: истинные значения
    :param y_pred: предсказанные значения
    :param tolerance: допустимая относительная ошибка (например, 0.1 = 10%)
    :return: точность (accuracy)
    """
    errors = np.abs((y_pred - y_true) / y_true)
    accuracy = np.mean(errors <= tolerance)
    return accuracy

# Пример использования для трех моделей
metrics = []

# 1. Линейная регрессия
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)
r2_linear, rmse_linear = evaluate_model(y_test, y_pred_linear)
accuracy_linear = regression_accuracy(y_test, y_pred_linear, tolerance=0.1)
metrics.append(('Linear Regression', r2_linear, rmse_linear, accuracy_linear))

# 2. Дерево решений
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)
r2_tree, rmse_tree = evaluate_model(y_test, y_pred_tree)
accuracy_tree = regression_accuracy(y_test, y_pred_tree, tolerance=0.1)
metrics.append(('Decision Tree', r2_tree, rmse_tree, accuracy_tree))

# 3. Градиентный бустинг (XGBoost)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
r2_xgb, rmse_xgb = evaluate_model(y_test, y_pred_xgb)
accuracy_xgb = regression_accuracy(y_test, y_pred_xgb, tolerance=0.1)
metrics.append(('XGBoost', r2_xgb, rmse_xgb, accuracy_xgb))

# Вывод результатов
print(f"{'Model':<20} {'R2':<10} {'RMSE':<10} {'Accuracy':<10}")
print("-" * 50)
for model_name, r2, rmse, accuracy in metrics:
    print(f"{model_name:<20} {r2:<10.4f} {rmse:<10.4f} {accuracy:<10.4f}")

# Вычисление средних значений
average_r2 = np.mean([metric[1] for metric in metrics])
average_rmse = np.mean([metric[2] for metric in metrics])
average_accuracy = np.mean([metric[3] for metric in metrics])

print("\nAverage Metrics:")
print(f"Average R2: {average_r2:.4f}")
print(f"Average RMSE: {average_rmse:.4f}")
print(f"Average Accuracy: {average_accuracy:.4f}")
```

```python
# Функция для нормализации столбцов DataFrame
def normalize_dataframe(df, columns=None):
    """
    Нормализует значения в указанных столбцах DataFrame в диапазон [0, 1].
    
    :param df: pandas DataFrame
    :param columns: Список столбцов для нормализации. Если None, нормализуются все числовые столбцы.
    :return: DataFrame с нормализованными значениями
    """
    df_normalized = df.copy()  # Создаем копию DataFrame, чтобы не изменять исходный
    if columns is None:
        # Если список столбцов не указан, выбираем все числовые столбцы
        columns = df.select_dtypes(include=['number']).columns
    
    for col in columns:
        df_normalized[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
    
    return df_normalized
```
